```python
# Answer 2
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.xlabel("x-Axis")
plt.ylabel("Y-Axis")
plt.title("Scatter plot")
plt.show()

```

```python
# Answer 3
# using subplot in Matplotlib enhances the flexibility and effectiveness of data visualization, 
# allowing you to present and analyze multiple plots in a single figure conveniently.

x = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])
y2 = np.array([50,20,40,20,60,70])
y3 = np.array([10,20,30,40,50,60])
y4 = np.array([200,350,250,550,450,150])

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
# Answer 4
# A Matplotlib bar plot is a type of plot that is used to visualize categorical data.
# It is made up of a series of rectangular bars, each of which represents the value of a category
# Matplotlib bar plots are used to compare the values of different categories. 
# For example, you could use a bar plot to compare the number of people who voted for each candidate in an election.
# Bar plots can also be used to show the distribution of data.
# For example, you could use a bar plot to show the distribution of test scores.
company = np.array(["Apple","Microsoft","Google","AMD"])
profit = np.array([3000,8000,1000,10000])

plt.bar(company,profit)
plt.show()

plt.barh(company,profit)
plt.show()
```
